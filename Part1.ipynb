```python
# Import essential libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Display all columns (not just some)
pd.set_option('display.max_columns', None)
```

```python
# load our car prices data
df = pd.read_csv('car_prices.csv')
print(df.shape)
df.head()
```

```python
# basic exploration
df.info()
print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# look at first rows
df.head()
# get basic statistics
df.describe()
# check data types
df.dtypes
```

```python
# Calculate the average selling price for each model
avg_selling_price_by_model = df.groupby('model')['sellingprice'].mean().sort_values(ascending=False)

# Select the top 10 models
top_10_models = avg_selling_price_by_model.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
top_10_models.plot(kind='bar')
plt.title('Average Selling Price for Top 10 Car Models')
plt.xlabel('Model')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```
